# Variational Autoencoder

VAE are models learning to reconstruct input data from a latent space. The model is made of two parts:
- The encoder part, which learns to encode the input data into a latent space
- The decoder part, which reconstructs the input data from the latent space

A constraint is set on the latent space in order to make it possible to easily generate new data. The latent space is constrained to follow Gaussian distribution.
